# 함수로 데이터 변경하기

```python
df =  pd.read_csv('data/scores.csv')
df = df.head()
df_copy = df.copy()
```

```python
df.head()
```

```python
df_copy.math = df.math+5
df_copy
```

```python
def plus5(x):
    score = x+5
    
    if score>=100:
        score=100
    
    return score

df_copy['math'] = df['math'].apply(plus5)

df_copy
```

```python
def plusn(x,n):
    score = x+n
    
    if score>=100:
        score=100
    
    return score

df_copy['eng'] = df['eng'].apply(plusn,n=1)
df_copy
```

```python
df =  pd.read_csv('data/scores.csv')
df = df.head()
df.index = df.name
df.drop(columns=['name'], inplace=True)
df_copy = df.copy()
df_copy
```

## 행단위 데이터 출력

```python
df_copy['sum'] = df.apply(get_sum,axis=1)
```

```python
df_copy.loc['sum'] = df.apply(get_sum, axis=0)
df_copy
```

# 결측치 처리하기

결측치(Null) : 공백이 아닌 값이없다는 뜻

```python
import numpy as np
df = pd.DataFrame([[np.nan, 2, np.nan,0],
                   [3,4,np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan,4]],
                  columns=list('ABCD'))
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

# 결측치 삭제하기

```python
# 결측치가 존재하는 모든 행 삭제

df.dropna()
```

```python
# 결측치가 존재하는 모든 열 삭제

df.dropna(axis=1)
```

# 결측치 대치하기


```python
df
```

```python
# 0으로 채우기

df.fillna(0)
```

```python
# 평균값으로 채우기(컬럼별 평균값으로 채워진다.)

df.fillna(df.mean())
```

```python
# 이전값으로 채우기

df.fillna(method='ffill')
```

```python
# 다음값으로 채우기

df.fillna(method='bfill')
```

```python

# 컬럼별로 값을 지정해서 채우기

df.fillna({'A':0, 'B':1, 'C':2,'D':3})
```
